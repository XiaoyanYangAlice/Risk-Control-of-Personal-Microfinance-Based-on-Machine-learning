```python
import pandas as pd

f1 = open(r'C:\Users\XPS\Desktop\数据\LoanStats_securev1_2019Q1.csv',encoding='gb18030',errors='ignore')
loans_1901 = pd.read_csv(f1,skiprows=1)

##缺失值处理，将阈值设在50%，先清洗掉缺失严重的特征指标
half_count = len(loans_1901) / 2
loans_1901 = loans_1901.dropna(thresh=half_count, axis=1)
```

```python
##进行第一次特征选择
col_n = ["loan_amnt", "term", "int_rate","installment","emp_length", "home_ownership", "annual_inc", "verification_status","loan_status","pymnt_plan","purpose","addr_state","dti","delinq_2yrs","inq_last_6mths","open_acc", "pub_rec","revol_bal", "revol_util", "total_acc" ,"initial_list_status", "collections_12_mths_ex_med" ,"application_type", "acc_now_delinq","chargeoff_within_12_mths","delinq_amnt","pub_rec_bankruptcies","tax_liens"]
loans_1901 = pd.DataFrame(loans_1901,columns = col_n)

print(loans_1901.iloc[0])
print(loans_1901.shape[1])

print(loans_1901['loan_status'].value_counts())
```

```python
##保存一下
loans_1901.to_csv(r'C:\Users\XPS\Desktop\数据\loans_1901.csv', index=False)
```

```python
##进行Y的编码
loans_1901 = loans_1901[(loans_1901['loan_status'] == "Fully Paid") | 
                        (loans_1901['loan_status'] == "Charged Off")]

status_replace = {
    "loan_status" : {
        "Fully Paid": 1,
        "Charged Off": 0,
    }
}

loans_1901 = loans_1901.replace(status_replace)
```

```python
##寻找一下只有一个取值（即方差为零）的特征指标并删除，再次保存数据
orig_columns = loans_1901.columns
drop_columns = []
##需要先去掉缺失值再看
for col in orig_columns:
    col_series = loans_1901[col].dropna().unique()
    if len(col_series) == 1:
        drop_columns.append(col)
loans_1901 = loans_1901.drop(drop_columns, axis=1)
print(drop_columns)
print(loans_1901.shape)
loans_1901.to_csv(r'C:\Users\XPS\Desktop\数据\filtered_loans_1901.csv', index=False)  ##至此保存好了训练数据集
```

```python
##整理当前数据集的数据类型，下面将对object类型数进行数据类型转换
##要是后面有问题从这一条开始重新运行#########################################
loans = loans_1901.dropna(axis=0)
print(loans.dtypes.value_counts())

object_columns_df = loans.select_dtypes(include=["object"])   ##查看非数值特征取值样例
print(object_columns_df.iloc[0])
```

```python
##统计各个取值占比（可以画图）
cols = ['home_ownership', 'verification_status', 'emp_length', 'term', 'addr_state']
for c in cols:
    print(loans[c].value_counts())
print(loans["purpose"].value_counts())
```

```python
mapping_dict = {
    "emp_length": {
        "10+ years": 10,
        "9 years": 9,
        "8 years": 8,
        "7 years": 7,
        "6 years": 6,
        "5 years": 5,
        "4 years": 4,
        "3 years": 3,
        "2 years": 2,
        "1 year": 1,
        "< 1 year": 0,
        "n/a": 0
    }
}
#loans = loans.drop(["last_credit_pull_d", "earliest_cr_line", "addr_state"], axis=1)

loans = loans.replace(mapping_dict)
```

```python
loans["term"] = loans["term"].astype(str).str.strip("months").astype("int")
loans["int_rate"] = loans["int_rate"].astype(str).str.strip("%").astype("float")
loans["revol_util"] = loans["revol_util"].astype(str).str.strip("%").astype("float")

loans = loans.drop("addr_state", axis=1)
loans = loans.drop("initial_list_status", axis=1)
```

```python
import pandas as pd

f2 = open(r'C:\Users\XPS\Desktop\数据\loans_1901_cleaned.csv',encoding='gb18030',errors='ignore')
loans = pd.read_csv(f2)
print(loans.info())
```

```python
from sklearn.linear_model import LogisticRegression

#loans = pd.read.csv(r'C:\Users\mlamp\Desktop\数据\loans_1901_cleaned.csv',)
cols = loans.columns
model_cols = cols.drop("loan_status")
X = loans[model_cols]
y = loans["loan_status"]
```

```python
# 构建自变量和因变量
n_sample = y.shape[0]
n_pos_sample = y[y == 1].shape[0]
n_neg_sample = y[y == 0].shape[0]
print('样本个数：{}; 正样本占{:.2%}; 负样本占{:.2%}'.format(n_sample,
                                                   n_pos_sample / n_sample,
                                                   n_neg_sample / n_sample))
print('特征维数：', X.shape[1])

# 处理不平衡数据
from imblearn.over_sampling import SMOTE# 使用imlbearn库中上采样方法中的SMOTE接口
sm = SMOTE(random_state=42)    # 处理过采样的方法
X, y = sm.fit_resample(X, y)
print('通过SMOTE方法平衡正负样本后')
n_sample = y.shape[0]
n_pos_sample = y[y == 1].shape[0]
n_neg_sample = y[y == 0].shape[0]
print('样本个数：{}; 正样本占{:.2%}; 负样本占{:.2%}'.format(n_sample,
                                                   n_pos_sample / n_sample,
                                                   n_neg_sample / n_sample))
```

```python
#切割训练集 7:3划分
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0) # random_state = 0 每次切分的数据都一样
```

```python
##标准化数据集（作为对照组）
from sklearn.preprocessing import StandardScaler as SS
X_train_std=SS().fit_transform(X_train)
X_test_std=SS().fit_transform(X_test)
```

```python
##逻辑回归调参

import time 
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report


# 构建参数组合
param_grid_lr = {'C': [0.01,0.1, 1, 10, 100, 1000],
                            'penalty': [ 'l1', 'l2']}
# C：Inverse of regularization strength; must be a positive float. 
# Like in support vector machines, smaller values specify stronger regularization.

grid_search_lr = GridSearchCV(LogisticRegression(),param_grid_lr, cv=10) 
# 确定模型LogisticRegression，和参数组合param_grid ，cv指定10折
grid_search_lr.fit(X_train, y_train) # 使用训练集学习算法

#start = time.clock()

#elapsed = (time.clock() - start)


print("Best parameters: {}".format(grid_search_lr.best_params_))
print("Best cross-validation score: {:.5f}".format(grid_search_lr.best_score_))
print("Best estimator:\n{}".format(grid_search_lr.best_estimator_)) 


# grid_search.best_estimator_ 返回模型以及他的所有参数（包含最优参数）并验证 拟合效果
start = time.clock()
y_pred_lr = grid_search_lr.predict(X_test)
elapsed = (time.clock() - start)
print("Logistic Time used:",elapsed)

from sklearn.metrics import accuracy_score
print("Test set accuracy score: {:.5f}".format(accuracy_score(y_test, y_pred_lr,)))

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_curve, roc_auc_score

roc_auc = roc_auc_score(y_test, y_pred_lr)
print("Area under the ROC curve : %f" % roc_auc)
```

```python
##TPR　FPR
predictions = pd.Series(y_pred_lr)
# False positives.
fp_filter = (predictions == 0) & (loans["loan_status"] == 1)
fp = len(predictions[fp_filter])

# True positives.
tp_filter = (predictions == 1) & (loans["loan_status"] == 1)
tp = len(predictions[tp_filter])

# False negatives.
fn_filter = (predictions == 1) & (loans["loan_status"] == 0)
fn = len(predictions[fn_filter])

# True negatives
tn_filter = (predictions == 0) & (loans["loan_status"] == 0)
tn = len(predictions[tn_filter])

# Rates
tpr = tp / float((tp + fn))
fpr = fp / float((fp + tn))

print("TPR : %f" % tpr)

print("FPR : %f" % fpr)
```

```python
##随机森林模型
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

#rf = RandomForestClassifier(n_estimators=10,class_weight="balanced", random_state=1)
rf = RandomForestClassifier()


parameter_grid_rf = {'n_estimators': [5,10,15,20],
                  'criterion': ['gini', 'entropy'],
                  'max_features': [5,10,15,20],
                  'warm_start': [True, False]}

grid_search_rf = GridSearchCV(rf,
                           param_grid=parameter_grid_rf,
                           cv=10)

grid_search_rf.fit(X_train, y_train)


print("Best parameters: {}".format(grid_search_rf.best_params_))
print("Best cross-validation score: {:.5f}".format(grid_search_rf.best_score_))
print("Best estimator:\n{}".format(grid_search_rf.best_estimator_)) 
# grid_search.best_estimator_ 返回模型以及他的所有参数（包含最优参数）
start= time.clock()
y_pred_rf = grid_search_rf.predict(X_test)
elapsed = (time.clock() - start)
print("RF Time used:",elapsed)


##test accuracy 和 ROC
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import accuracy_score
print("Test set accuracy score: {:.5f}".format(accuracy_score(y_test, y_pred_rf,)))
roc_auc = roc_auc_score(y_test, y_pred_rf)
print("Area under the ROC curve : %f" % roc_auc)
```
